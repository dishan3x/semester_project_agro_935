# Challenge: Generate a Complementary DNA Sequence

```python
dna = open('dna_sequence.txt').read().split('\n')[8:]
#print(dna)
```

```python
longString = ""
for danCharacterNum in range(0, len(dna)):  
    longString  = longString + dna[danCharacterNum]
    
#print(longString)
```

```python
newDNASequence = ""
for danCharacterNum in range(0, len(longString)):  
    dnaCharacter=  longString[danCharacterNum]
    #[A-T] and cytosine-guanine [C-G]
    if dnaCharacter == 'A' :
        #print("T")
        newDNASequence = newDNASequence + "T"
    elif dnaCharacter == 'T' :
        #print("A")
        newDNASequence = newDNASequence + "A"
    elif dnaCharacter == 'C' :
        #print("G")
        newDNASequence = newDNASequence + "G"
    elif dnaCharacter == 'G' :
        #print("C")
        newDNASequence = newDNASequence + "C"
    else:
        print("Error in the sequence")
        break
    
```
