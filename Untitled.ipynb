```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as  np
```

```python
df = pd.read_csv('dataSets/mauna_loa_co2.csv')
df.head()
```

```python
df.tail()
```

```python
df['month']
```

```python
df['decimal_date'] = df['year'] + df['month']/12
df
```

```python
plt.figure(figsize=(6,4),dpi=300)
plt.plot(df["decimal_date"],df["co2_ppm"])
plt.show()
```

```python
#print(plt.rcParams.items)
idx_2010s = df["year"] >= 2010
df_2010s = df[idx_2010s]
df_2010s.head()
```

```python
plt.figure()
plt.plot(df_2010s['decimal_date'],df_2010s['co2_ppm'])
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.tight_layout(); #remove overlapping and tighten it for a subpot
```
